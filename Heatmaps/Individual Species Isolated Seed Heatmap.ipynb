```python
def parse_fas_to_dict(filename):
    result = {}
    with open(filename, 'r') as f:
        label = None
        for line in f:
            line = line.strip()
            if line.startswith('>'):
                label = line[1:4]  # Get first 3 letters after '>'
            elif label:
                substring = line[1:8]  # Extract str[1:8]
                substring = substring.replace('U','T')
                if label not in result:
                    result[label] = []
                result[label].append(substring)
    return result

def count_unique_sequences(data_dict):
    unique_dict = {}
    shared_dict = {}

    # Flatten all sequences into a key-to-set mapping
    key_to_set = {k: set(v) for k, v in data_dict.items()}

    for key, seq_set in key_to_set.items():
        # Union of all other sets
        other_seqs = set().union(*(s for k, s in key_to_set.items() if k != key))
        # Sequences that are only in this key
        unique_seqs = seq_set - other_seqs
        unique_dict[key] = set(unique_seqs)
        shared_dict[key] = set(seq_set-unique_seqs)

    return unique_dict, shared_dict

def update_keys_with_species_names(d, species_filename):
    updated_dict = {}
    unmatched_keys = []
    used_abbrevs = set()
    abbrev_to_species = {}
    
    # Load species lines
    with open(species_filename, 'r') as f:
        lines = [line.strip() for line in f if line.strip()]
    
    # Build lookup from abbreviation (like 'hsa') to full species line
    for line in lines:
        if '(' in line and ')' in line:
            try:
                inner = line.split('(')[1].split(')')[0]
                words = inner.strip().split()
                if len(words) >= 2:
                    abbrev = words[0][0].lower() + words[1][:2].lower()
                    if abbrev in abbrev_to_species:
                        raise ValueError(f"Duplicate species abbreviation found: '{abbrev}' for '{line}'")
                    abbrev_to_species[abbrev] = line
            except IndexError:
                continue  # Skip malformed lines
    
    # Replace keys in original dict
    for key, value in d.items():
        abbrev = key.lower()
        if abbrev in abbrev_to_species:
            new_key = abbrev_to_species[abbrev]
            updated_dict[new_key] = value
            used_abbrevs.add(abbrev)
        else:
            unmatched_keys.append(key)
    
    # Get unmatched species lines
    unmatched_species_lines = [
        line for abbr, line in abbrev_to_species.items() if abbr not in used_abbrevs
    ]
    
    return updated_dict, unmatched_keys, unmatched_species_lines

# Example usage
filename = 'data/ALL-mature.fas'
d = parse_fas_to_dict(filename)
unique, shared = count_unique_sequences(d)
unique_with_labels, unmatched_keys, unmatched_species = update_keys_with_species_names(unique, "data/names.txt")
shared_with_labels, unmatched_keys, unmatched_species = update_keys_with_species_names(shared, "data/names.txt")
print(unmatched_keys)
print(unmatched_species)
```
